```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sum_stat = pd.read_csv('results/fitness_distribution/max/stat/all_lib_stats.csv')
sum_stat
```

```python
import ast
```

```python
lib_basic_df = pd.concat(
    [
        sum_stat["lib"],
        sum_stat["lib_basic_dict"].apply(ast.literal_eval).apply(pd.Series),
    ],
    axis=1,
)
lib_basic_df.style.background_gradient(cmap="YlGnBu")
```

```python
lib_basic_df["parnet_percent_rank"] = lib_basic_df["parent_rank"]/lib_basic_df["numb_measured"]*100
lib_basic_df
```

```python
lib_basic_df[
    [
        "lib",
        "n_site",
        "percent_measured",
        "percent_active",
        "active_fit_min",
        "parnet_percent_rank",
    ]
].style.background_gradient(cmap="YlGnBu").format({
    'n_site': '{:.0f}',
    'percent_measured': '{:.2f}',
    'percent_active': '{:.2f}',
    'active_fit_min': '{:.2f}',
    'parnet_percent_rank': '{:.2f}',
})
```

```python
fit_df = pd.concat(
    [
        sum_stat["lib"],
        sum_stat["fit_basic_dict"].apply(ast.literal_eval).apply(pd.Series),
    ],
    axis=1,
)
# Assuming each list in 'A' has the same length
expanded_df = pd.DataFrame(fit_df['quartiles'].to_list(), index=fit_df.index)

# Assign the new column names
expanded_df.columns = [f'q{i+1}' for i in range(len(expanded_df.columns))]

# join
fit_df = fit_df.drop(columns=["quartiles"]).join(expanded_df)

fit_df.style.background_gradient(cmap="YlGnBu")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [3, 4, 1, 5, 2],
    'C': [5, 3, 4, 1, 2]
})

# Apply a colormap to the DataFrame
styled_df = df.style.background_gradient(cmap='viridis')

# Render the DataFrame as a Matplotlib table
fig, ax = plt.subplots(figsize=(12, 2))  # set size frame
ax.axis('tight')
ax.axis('off')
the_table = table(ax, styled_df.data, loc='center', colWidths=[0.17]*len(df.columns))

# Save the table as a PNG image
# plt.savefig('df_styled.png', dpi=200, bbox_inches='tight')
```
