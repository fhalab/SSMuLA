```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sum_stat = pd.read_csv('results/fitness_distribution/max/all_lib_stats.csv')
sum_stat
```

```python
import ast
```

```python
pd.concat(
    [
        sum_stat["lib"],
        sum_stat["n_mut_cuttoff"],
        sum_stat["cauchy"].apply(ast.literal_eval).apply(pd.Series),
        sum_stat["kde"].apply(ast.literal_eval).apply(pd.Series),
    ],
    axis=1,
)
```

```python
from SSMuLA.landscape_global import n_mut_cutoff_dict
```

```python
for n in [0, 1, 2]:

    df = sum_stat[sum_stat["n_mut_cuttoff"] == n]

    print(n_mut_cutoff_dict[n])

    lib_basic_df = pd.concat(
        [
            df["lib"],
            # df["n_mut_cuttoff"],
            df["lib_basic_dict"].apply(ast.literal_eval).apply(pd.Series),
        ],
        axis=1,
    )
    lib_basic_df["parnet_percent_rank"] = (
        lib_basic_df["parent_rank"] / lib_basic_df["numb_measured"] * 100
    )

    display(lib_basic_df.style.background_gradient(cmap="YlGnBu"))
```

```python
lib_basic_df[
    [
        "lib",
        "n_site",
        "percent_measured",
        "percent_active",
        "active_fit_min",
        "parnet_percent_rank",
    ]
].style.background_gradient(cmap="YlGnBu").format({
    'n_site': '{:.0f}',
    'percent_measured': '{:.2f}',
    'percent_active': '{:.2f}',
    'active_fit_min': '{:.2f}',
    'parnet_percent_rank': '{:.2f}',
})
```

```python
fit_df = pd.concat(
    [
        sum_stat["lib"],
        sum_stat["fit_basic_dict"].apply(ast.literal_eval).apply(pd.Series),
    ],
    axis=1,
)
# Assuming each list in 'A' has the same length
expanded_df = pd.DataFrame(fit_df['quartiles'].to_list(), index=fit_df.index)

# Assign the new column names
expanded_df.columns = [f'q{i+1}' for i in range(len(expanded_df.columns))]

# join
fit_df = fit_df.drop(columns=["quartiles"]).join(expanded_df)

fit_df.style.background_gradient(cmap="YlGnBu")
```
