```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
old = pd.read_csv("results/mlde/all_df_comb_onehot_4.csv")
new = pd.read_csv("results/mlde_TrpB3F/vis/all_df.csv")
```

```python
new_sliced = new[(new["lib"]=='TrpB3F') & (new["model"]=="boosting") & (new["rep"].isin(range(50)))]
```
