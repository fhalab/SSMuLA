```python
def get_val_percent(s: pd.Series, numb: float = 1) -> float:
    """
    Get the Percent of a column that is equal to a certain number.
    """
    return (s == numb).sum().sum() / len(s) * 100
```

```python
# combine all in one
# Create the figure and subplots

plot_folder = checkNgen_folder("results/de_vs_mlde/onehot/collage/n_mean_frac")
mlde_all = pd.read_csv("results/mlde/vis_4/all_df.csv").copy()

for zs, mlde_title in zip(
    ["Triad_score", "esm_score", "ev_score", "none"],
    ["Triad-ftMLDE", "ESM-ftMLDE", "EVmutation-ftMLDE", "MLDE"],
):

    sup_title = f"{mlde_title} vs DE over sample size"

    ncol = 5
    nrow = 3
    fig, axs = plt.subplots(nrow, ncol, figsize=(40, 16), sharex=True, sharey=True)

    for i, (ax, lib) in enumerate(zip(axs.flatten(), LIB_INFO_DICT.keys())):

        ss_df_all = pd.read_csv(
            f"results/de/DE-active/scale2max/{lib}-single_step_DE.csv"
        )
        recomb_df_all = pd.read_csv(
            f"results/de/DE-active/scale2max/{lib}-recomb_SSM.csv"
        )

        ss_df = ss_df_all.copy()
        recomb_df = recomb_df_all.copy()

        ss_mean = ss_df["final_fitness"].mean()
        ss_frac = get_val_percent(s=ss_df["final_fitness"], numb=1)

        recomb_mean = recomb_df["final_fitness"].mean()
        recomb_frac = get_val_percent(s=recomb_df["final_fitness"], numb=1)

        mlde_df = mlde_all[
            (mlde_all["lib"] == lib)
            & (mlde_all["n_mut_cutoff"] == "all")
            & (mlde_all["zs"] == zs)
            & (mlde_all["encoding"] == "one-hot")
        ].copy()

        mean_fit_dict = {}
        max_percent_dict = {}

        for n_samples in [24, 48, 96, 192, 288, 384, 480, 576, 960, 1920]:
            # [24, 48, 96, 192, 288, 384, 480, 576, 960, 1920]
            mlde_df_n = mlde_df[mlde_df["n_sample"] == n_samples]["top_maxes"]

            mean_fit_dict[n_samples] = mlde_df_n.mean()
            max_percent_dict[n_samples] = get_val_percent(
                s=mlde_df_n, numb=1
            )

        x = list(mean_fit_dict.keys())
        y1 = list(mean_fit_dict.values())
        y2 = list(max_percent_dict.values())

        # Plotting the first y-axis
        ax.plot(x, y1, color=ten_colors[0], marker="o", linewidth=2, label="Mean of max fitness: MLDE")
        ax.axhline(
            y=ss_mean,
            color=ten_colors[0],
            linestyle="dotted",
            linewidth=2,
            label="Mean of max fitness: Single step",
        )
        ax.axhline(
            y=recomb_mean,
            color=ten_colors[0],
            linestyle="dashed",
            linewidth=2,
            label="Mean of max fitness: Recomb",
        )

        # Setting labels and title
        ax.set_xlabel("Training set size")
        ax.set_ylabel("Mean of max fitness")
        ax.tick_params(axis="y")

        # Creating a second y-axis
        ax2 = ax.twinx()
        ax2.plot(
            x, y2, color=ten_colors[-1], marker="s", linewidth=2, label="Percent of max fitness = 1: MLDE"
        )
        ax2.axhline(
            y=ss_frac,
            color=ten_colors[-1],
            linestyle="dotted",
            linewidth=2,
            label="Percent of final fitness = 1: single step",
        )
        ax2.axhline(
            y=recomb_frac,
            color=ten_colors[-1],
            linestyle="dashed",
            linewidth=2,
            label="Percent of final fitness = 1: recomb",
        )

        ax2.set_ylabel("Percent of max fitness = 1")
        ax2.tick_params(axis="y")

        if i == ncol - 1:
            ax.legend(loc="upper left", bbox_to_anchor=(1.15, 1))
            ax2.legend(loc="upper left", bbox_to_anchor=(1.15, 0.5))

        ax.set_title(lib)
        ax.set_ylim(0, 1)
        ax2.set_ylim(0, 100)


    fig.suptitle(sup_title, fontsize=16, fontweight="bold", y=0.9125)

    save_plt(
        fig,
        plot_title=sup_title,
        path2folder=plot_folder,
    )
```

```python
def plot_de_mlde_ft_v_n(
    plot_folder: str = "results/de_vs_mlde/onehot/collage/n_mean_frac",
    mlde_csv: str = "results/mlde/vis_4/all_df.csv",
    de_folder: str = "results/de/DE-active/scale2max",
) -> None:
    """"""
    # combine all in one
    # Create the figure and subplots

    plot_folder = checkNgen_folder(plot_folder)
    mlde_all = pd.read_csv(mlde_csv).copy()

    sup_title = "DE vs MLDE over sample size"

    ncol = 5
    nrow = 3
    fig, axs = plt.subplots(nrow, ncol, figsize=(40, 16))

    for i, (ax, lib) in enumerate(zip(axs.flatten(), LIB_INFO_DICT.keys())):
        print(i, ax, lib)

        ss_df = pd.read_csv(f"{de_folder}/{lib}-single_step_DE.csv").copy()
        recomb_df = pd.read_csv(f"{de_folder}/{lib}-recomb_SSM.csv").copy()

        ss_mean = ss_df["final_fitness"].mean()
        ss_frac = get_val_percent(s=ss_df["final_fitness"], numb=1)

        recomb_mean = recomb_df["final_fitness"].mean()
        recomb_frac = get_val_percent(s=recomb_df["final_fitness"], numb=1)

        print("hi there")

        for (zs_label, zs_color, zs) in ZS_COLOR_ZIP:

            print(zs_label, zs_color, zs)

            mean_fit_dict = {}
            max_percent_dict = {}

            for n_samples in N_SAMPLE_LIST:

                print(n_sample)

                mlde_df = mlde_all[
                    (mlde_all["lib"] == lib)
                    & (mlde_all["n_mut_cutoff"] == "all")
                    & (mlde_all["zs"] == zs)
                    & (mlde_all["encoding"] == "one-hot")
                ].copy()

                print(mlde_df)

                if zs == "none":
                    mlde_df_n = mlde_df[(mlde_df["zs"] == zs) & (mlde_df["n_sample"] == n_samples)]["top_maxes"]

                else:
                    mlde_df_n = mlde_df[
                        (mlde_df["zs"] == zs)
                        & (mlde_df["ft_lib"] == mlde_df["ft_lib"].min())
                        & (mlde_df["n_sample"] == n_samples)
                    ]["top_maxes"]

                print(mlde_df_n)

                mean_fit_dict[n_samples] = mlde_df_n.mean()
                max_percent_dict[n_samples] = get_val_percent(s=mlde_df_n, numb=1)

            x = list(mean_fit_dict.keys())
            y1 = list(mean_fit_dict.values())
            y2 = list(max_percent_dict.values())

            print(x, y1, y2)

            # Plotting the first y-axis
            ax.plot(
                x,
                y1,
                color=zs_color,
                marker="o",
                linestyle="solid",
                linewidth=2,
                label="Mean of final fitness: MLDE",
            )
            ax.axhline(
                y=ss_mean,
                color=PRESENTATION_PALETTE_SATURATE["orange"],
                linestyle="dotted",
                label="Mean of final fitness: single step",
            )
            ax.axhline(
                y=recomb_mean,
                color=PRESENTATION_PALETTE_SATURATE["yellow"],
                linestyle="dashed",
                label="Mean of final fitness: recomb",
            )

            # Setting labels and title
            ax.set_xlabel("Training set size")
            ax.set_ylabel("Mean of max fitness")
            ax.tick_params(axis="y")

            # Creating a second y-axis
            ax2 = ax.twinx()
            ax2.plot(
                x,
                y2,
                color=zs_color,
                marker="s",
                linestyle="dashed",
                linewidth=2,
                label="Fraction of final fitness = 1: MLDE",
            )
            ax2.axhline(
                y=ss_frac,
                color=PRESENTATION_PALETTE_SATURATE["orange"],
                linestyle="dotted",
                label="Fraction of final fitness = 1: single step",
            )
            ax2.axhline(
                y=recomb_frac,
                color=PRESENTATION_PALETTE_SATURATE["yellow"],
                linestyle="dashed",
                label="Fraction of final fitness = 1: recomb",
            )

            ax2.set_ylabel("Fraction of max fitness = 1")
            ax2.tick_params(axis="y")

            if i == ncol - 1:
                # create labels
                ax.legend(loc="upper left", bbox_to_anchor=(1.12, 1))
                ax2.legend(loc="upper left", bbox_to_anchor=(1.12, 0.5))

            ax.set_title(lib)
            ax.set_ylim(0, 1)
            ax2.set_ylim(0, 100)

    fig.suptitle(sup_title, fontsize=16, fontweight="bold", y=0.9125)

    save_plt(
        fig,
        plot_title=sup_title,
        path2folder=plot_folder,
    )
```
