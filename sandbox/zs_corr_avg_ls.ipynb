```python
ZS_OPTS_LEGEND_append = {
    "ed_Triad-ev_score": "Edited distance + Triad + EVmutation", 
    "ed_Triad-esm_score": "Edited distance + Triad + ESM", 
    "ed_struc-comb_score": "Edited distance + Triad + ESM-IF", 
    "ed_msanoif-comb_score": "Edited distance + EVmutation + ESM", 
    "ed_msa-comb_score": "Edited distance + EVmutation + ESM + ESM-IF", 
    "ed_structnmsa-comb_score": "Edited distance + Triad + EVmutation + ESM + ESM-IF", 
}
```

```python
corr_name = ZS_OPTS+ZS_COMB_OPTS+["ed_Triad-ev_score", "ed_Triad-esm_score", "ed_struc-comb_score", "ed_msanoif-comb_score","ed_msa-comb_score", "ed_structnmsa-comb_score"]

corr_dict = {}
for lib in ['DHFR', 'ParD2', 'ParD3', 'GB1', 'TrpB3D', 'TrpB3E', 'TrpB3F',
       'TrpB3G', 'TrpB3I', 'TrpB4']:

    lib_csv = f"results/zs_comb_2/none/scale2max/all/{lib}.csv"
    # ds_lib_csv = f"results/zs_comb_2/none/scale2max/double/{lib}.csv"

    df = pd.read_csv(lib_csv)
    # ds_df = pd.read_csv(ds_lib_csv)

    df["ed_Triad-ev_score"] = -1 * (df["Triad_rank"] + df["ev_rank"]+ df["ed_rank"])
    df["ed_Triad-esm_score"] = -1 * (df["Triad_rank"] + df["esm_rank"]+ df["ed_rank"])

    df["ed_struc-comb_score"] = -1 * (df["Triad_rank"] + df["esmif_rank"]+ df["ed_rank"])
    df["ed_msa-comb_score"] = -1 * (df["ev_rank"] + df["esm_rank"] + df["esmif_rank"]+ df["ed_rank"])
    df["ed_msanoif-comb_score"] = -1 * (df["ev_rank"] + df["esm_rank"] + df["ed_rank"])
    df["ed_structnmsa-comb_score"] = -1 * (
        df["Triad_rank"] + df["ev_rank"] + df["esm_rank"] + df["esmif_rank"] + df["ed_rank"]
    )
    corr_dict[lib] = df[corr_name].corr(method="spearman")

avg_corr_df = pd.DataFrame(np.mean(np.stack(list(corr_dict.values())),axis=0),
    index=list(ZS_OPTS_LEGEND.values())[1:] + list(ZS_OPTS_LEGEND_append.values()),
    columns=list(ZS_OPTS_LEGEND.values())[1:] + list(ZS_OPTS_LEGEND_append.values())
)
# Create a mask for the upper triangle
mask = np.triu(np.ones_like(avg_corr_df, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(
    avg_corr_df,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="Blues",
    cbar_kws={"shrink": 0.6},
    linewidths=0.8,
    square=True,
    vmin=0,
    vmax=1,
    annot_kws={'size': 8},
)  # Setting vmin and vmax

# Add titles and labels as necessary
plt.title("ZS Correlations Average Across 10 Minimally Active Landscapes", fontweight="bold")
# plt.show()
```
