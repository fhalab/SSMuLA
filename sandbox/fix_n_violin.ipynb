```python

def plot_de_mlde_ft_fix_n(
    plot_folder: str = "results/de_vs_mlde/onehot/collage/n_mean_frac",
    mlde_csv: str = "results/mlde/all_df_comb.csv",
    de_folder: str = "results/de/DE-active/scale2max",
    meanorfrac: str = "mean",
    n_mut: str = "all",
    n_sample: int = 384,
    n_top: int = 96,
    min_ftlib: bool = True,
    simplezs: bool = True,
    addhd: bool = True,
    liborderby: str = "single_step_DE",
) -> None:
    """"""
    # combine all in one
    # Create the figure and subplots

    """
    Plot DE and MLDE with different ZS as violin plots for each library
    """

    ncol = 5
    nrow = 3
    fig, axs = plt.subplots(nrow, ncol, figsize=(32, 16), sharex=True, sharey=True)

    if min_ftlib:
        app_ftlib = " (12.5% ft)"
    else:
        app_ftlib = ""

    if meanorfrac == "mean":
        app_meanfrac = "mean of max fitness"
        y_label = "Mean of max fitness"
    else:
        app_meanfrac = "fraction of max fitness = 1"
        y_label = "Fraction of max fitness = 1"

    if simplezs:
        app_zs = ""
        zs_opts = ZS_OPTS
    else:
        app_zs = " with ZS ensemble"
        zs_opts = ZS_OPTS + ZS_COMB_OPTS

    if addhd:
        app_hd = " with HD"
        zs_opts = ["none"] + N_MUT_SUBS + zs_opts
    else:
        app_hd = ""
        zs_opts = ["none"] + zs_opts

    plot_folder = checkNgen_folder(
        os.path.join(
            "".join([plot_folder, app_zs.replace(" ", "_")]), n_mut, str(n_top)
        )
    )

    mlde_all = pd.read_csv(mlde_csv).copy()

    sup_title = (
        f"DE vs MLDE{app_ftlib} {n_sample} top{str(n_top)} {app_meanfrac} {app_hd}"
    )

    de_sumdf = pd.read_csv(f"{de_folder}/all_landscape_de_summary.csv")
    liborder = list(
        de_sumdf[de_sumdf["de_type"] == liborderby]
        .sort_values(by=["mean_all"])["lib"]
        .values
    )

    for i, (ax, lib) in tqdm(enumerate(zip(axs.flatten(), liborder))):

        df_lib_list = []

        # now add DE
        for de in DE_TYPES:
            de_df = pd.read_csv(f"{de_folder}/{lib}-{de}.csv").copy()
            de_df["zs"] = de
            df_lib_list.append(de_df.rename(columns={"final_fitness": "top_maxes"}))
            
            ax.axhline(
                y=de_df["final_fitness"].mean(),
                color=DE_COLORS[de],
                linestyle="dotted",
                linewidth=2,
                label=f"DE mean: {DE_LEGEND_MAP[de]}",
            )

        for zs in zs_opts:

            if zs in N_MUT_SUBS:
                # N_MUT_SUBS = ["double", "single"]
                df = mlde_all[
                    (mlde_all["lib"] == lib)
                    & (mlde_all["n_mut_cutoff"] == zs)
                    & (mlde_all["n_top"] == n_top)
                    & (mlde_all["encoding"] == "one-hot")
                    & (mlde_all["zs"] == "none")
                ].copy()

                df["zs"] = zs

            else:
                # Common condition
                df = mlde_all[
                    (mlde_all["lib"] == lib)
                    & (mlde_all["n_mut_cutoff"] == n_mut)
                    & (mlde_all["n_top"] == n_top)
                    & (mlde_all["encoding"] == "one-hot")
                    & (mlde_all["zs"] == zs)
                ].copy()

                # Apply the additional condition if needed
                if min_ftlib and zs != "none":
                    df = df[(df["ft_lib"] == df["ft_lib"].min())]

                if zs == "none":
                    # add random avg
                    ax.axhline(
                        y=df[df["n_sample"] == n_sample]["top_maxes"].mean(),
                        color=ZS_COLOR_MAP_APPENDED[zs],
                        linestyle="solid",
                        linewidth=2,
                        label="Random mean",
                    )

            df = df[df["n_sample"] == n_sample][["zs", "top_maxes"]]
            df_lib_list.append(df)

            # sns.violinplot(x='zs', y='top_maxes', hue='zs', data=mlde_all, palette=ZS_COLOR_MAP_APPENDED, ax=ax)

        sns.violinplot(x='zs', y='top_maxes', hue='zs', data=pd.concat(df_lib_list, axis=0), palette={**ZS_COLOR_MAP_APPENDED, **DE_COLORS}, ax=ax)
        
            # Plotting
            # ax.plot(
            #     list(fit_dict.keys()),
            #     list(fit_dict.values()),
            #     color=ZS_COLOR_MAP_APPENDED[zs],
            #     marker="o",
            #     linestyle="solid",
            #     linewidth=2,
            #     label=ZS_OPTS_APPENDED[zs],
            # )
            # sns.violinplot(x='Category', y='Value', hue='Group', data=df, palette=ZS_COLOR_MAP_APPENDED, ax=ax)

            

        # Setting labels and title
        ax.set_xlabel("Number of samples")
        ax.set_ylabel(y_label)
        ax.tick_params(axis="y")

        if i == ncol - 1:
            # create labels
            legend_list = []

            for de in DE_TYPES:
                legend_list.append(
                    Line2D(
                        [0],
                        [0],
                        color=DE_COLORS[de],
                        linewidth=2,
                        linestyle="dotted",
                        label=f"DE mean: {DE_LEGEND_MAP[de]}",
                    )
                )
            
            for zs in zs_opts:
                legend_list.append(mpatches.Patch(
                  
                    color=ZS_COLOR_MAP_APPENDED[zs],
                    label=ZS_OPTS_APPENDED[zs],
                )
                )

            ax.legend(handles = [legend_list], loc="upper left", bbox_to_anchor=(1.12, 1))

        ax.set_title(lib)
        ax.set_ylim(0, 1)

        extra_len = len(axs.flatten()) - len(liborder)

        if extra_len > 0:
            for i in range(extra_len):
                axs.flatten()[-(i + 1)].set_visible(False)

    fig.tight_layout
    fig.suptitle(sup_title, fontsize=16, fontweight="bold", y=0.9125)

    # save_plt(
    #     fig,
    #     plot_title=sup_title,
    #     path2folder=plot_folder,
    # )

```
