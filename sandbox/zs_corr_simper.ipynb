```python
# Define a function to apply the t-test
def apply_ttest(group, ):
    subgroup_x = group[group[subgroup] == 'Binding'][metricr]
    subgroup_y = group[group[subgroup] == 'Enzymatic activity'][metricr]
    if len(subgroup_x) > 1 and len(subgroup_y) > 1:  # Ensure enough data for t-test
        t_stat, p_value = ttest_ind(subgroup_x, subgroup_y, equal_var=False)
    else:
        t_stat, p_value = np.nan, np.nan  # Not enough data to perform t-test
    return pd.Series({
        't_stat': t_stat,
        'p_value': p_value,
        'Binding': np.mean(subgroup_x),
        'Enzymatic activity': np.mean(subgroup_y)
    })
```
