```python
import os
import pandas as pd
import numpy as np
from glob import glob
from scipy.stats import spearmanr
```

```python
from SSMuLA.util import get_file_name
```

```python
og_ev = glob("/disk2/fli/SSMuLA/ev_esm2/*/*.csv")
new_ev_dir = "/disk2/fli/REVIVAL2/SSMuLA-zs/ev/"
```

```python
import matplotlib.pyplot as plt
```

```python
for ev in og_ev:
    print(ev)
    df = pd.read_csv(ev)
    new_ev = pd.read_csv(os.path.join(new_ev_dir, get_file_name(ev)+".csv"))
    join_df = pd.merge(df[["muts", "n_mut", "ev_score"]], new_ev[["muts", "ev_score"]], on=["muts"], how="outer")
    # print nan values
    print(join_df[join_df.isnull().any(axis=1)])
    print(spearmanr(join_df["ev_score_x"], join_df["ev_score_y"], nan_policy="omit"))
    plt.scatter(join_df["ev_score_x"], join_df["ev_score_y"])
    plt.show()
```
