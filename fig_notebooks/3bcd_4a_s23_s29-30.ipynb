```python
import pandas as pd

from SSMuLA.zs_analysis import (
    plot_zs_metric_msa,
    plot_app_type_zs,
    plot_zs_corr
)

%matplotlib inline
```

```python
zs_parse_csv = "results/zs_sum/zs_stat_parsed.csv"
zs_df = pd.read_csv(zs_parse_csv)
lib_stat = pd.read_csv("results/landscape/lib_stats.csv")
active_cutoff = 1
active_lib_list = lib_stat[(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist() # & (lib_stat["percent_measured"]>=90)
active_lib_list
```

```python
plot_zs_metric_msa(
    lib_list=active_lib_list,
    zs_parse_csv=zs_parse_csv,
    fig_name="3bc",
    ifsave=True,
)
```

```python
plot_app_type_zs(
    metric="rho",
    n_mut="all",
    lib_list=active_lib_list,
    parse_csv=zs_parse_csv,
    fig_name = "4a",
    y_min = - 0.1,
    y_max = 0.65,
    y_annotation = 0.6,
    if_save = True,
)
```

```python
plot_app_type_zs(
    metric="rocauc",
    n_mut="all",
    lib_list=active_lib_list,
    parse_csv=zs_parse_csv,
    fig_name = "4a",
    y_min = 0.4,
    y_max = 1,
    y_annotation = 0.95,
    if_save = True,
)
```

```python
# for inactive
for m in ["rho", "rocauc"]:
    plot_app_type_zs(
        metric=m,
        n_mut="all",
        slice_zs=zs_df[
            zs_df["lib"].isin(
                lib_stat[~(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist()
            )
        ].copy(),
        # save_dir = "fig",
        fig_name="s23",
        if_save = True,
    )
```

```python
# for single mutation > 1% active
for m in ["rho", "rocauc"]:
    plot_app_type_zs(
        metric=m,
        n_mut="single",
        slice_zs=zs_df[zs_df["lib"].isin(active_lib_list)].copy(),
        y_annotation = 0.95,
        fig_name = "s29",
        # save_dir = "fig",
        if_save = True,
    )
```

```python
# for single mutation <1% active 
for m in ["rho", "rocauc"]:
    plot_app_type_zs(
        metric=m,
        n_mut="single",
        slice_zs=zs_df[~zs_df["lib"].isin(active_lib_list)].copy(),
        fig_name = "s30",
        # save_dir = "fig",
        if_save = True,
    )
```

```python
plot_zs_corr(
    n_mut_cutoff="all",
    active_cutoff=1,
    lib_stat_csv="results/landscape/lib_stats.csv", # path to the library statistics output csv 
    zs_df_dir="results/zs_comb", # path to the zs directory with all landscape zs score combined
    save_dir="figs",
    fig_id="3d",
    ifsave=True,
)
```
