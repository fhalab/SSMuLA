A notebook for regenerating plot for figure on different models for MLDE and ALDE performance

```python
import pandas as pd 
import numpy as np

from SSMuLA.mlde_analysis import get_model_comp_dict, plot_mlde_dict, plot_alde_opt

%matplotlib inline
```

```python
mlde_csv = "results/mlde/all_results.csv"
alde_dir = "results_rev/alde"
lib_stat = pd.read_csv("results/landscape/lib_stats.csv")
active_cutoff = 1
```

```python
# for landscapes with at least 1% active variants with boosting, ridge, alde boosting and DNN
plot_mlde_dict(
    get_model_comp_dict(
        mlde_csv,
        lib_stat[(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist(),
        n_top=96,
        alde_dir=alde_dir,
    ),
    fig_name="s37",
    n_top=96,
    n_corr=384,
    ifvline=False,
    fig_dir="figs",
    ifsave=True,
)
```

```python
# for landscapes with less than 1% active variants with boosting, ridge, alde boosting and DNN
plot_mlde_dict(
    get_model_comp_dict(
        mlde_csv,
        lib_stat[~(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist(),
        n_top=96,
        alde_dir=alde_dir,
    ),
    fig_name="s38",
    n_top=96,
    n_corr=384,
    ifvline=False,
    fig_dir="figs",
    ifsave=True,
)
```

```python
# for landscapes with at least 1% active variants with different ALDE opts
plot_alde_opt(
    alde_dir=alde_dir,
    lib_list=lib_stat[(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist(),
    fig_name="s39",
    n_top = 96,
    n_corr = 384,
    n_rounds = [2, 3, 4],
    ifzoomy = False,
    ifsave = True,
    fig_dir = "figs",
)
```

```python
# for landscapes with fewer than 1% active variants with different ALDE opts
plot_alde_opt(
    alde_dir=alde_dir,
    lib_list=lib_stat[(lib_stat["percent_active"] < active_cutoff)]["lib"].tolist(),
    fig_name="s40",
    n_top = 96,
    n_corr = 384,
    n_rounds = [2, 3, 4],
    ifzoomy = False,
    ifsave = True,
    fig_dir = "figs",
)
```
