A notebook for regenerating plot for figure S21 and S22 on different models for MLDE and ALDE performance

```python
import pandas as pd 
import numpy as np

from SSMuLA.mlde_analysis import get_model_comp_dict, plot_mlde_dict

%matplotlib inline
```

```python
mlde_csv = "/disk2/fli/SSMuLA/results4upload/mlde/one-hot.csv"
alde_dir = "/disk2/fli/alde4ssmula"
lib_stat = pd.read_csv("/disk2/fli/SSMuLA/results4upload/landscape/lib_stats.csv")
active_cutoff = 1
```

```python
# for landscapes with at least 1% active variants with boosting, ridge, alde boosting and DNN
plot_mlde_dict(
    get_model_comp_dict(
        mlde_csv,
        lib_stat[(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist(),
        n_top=96,
        alde_dir=alde_dir,
    ),
    fig_name="s21",
    n_top=96,
    n_corr=384,
    ifvline=False,
    fig_dir="figs",
    ifsave=True,
)
```

```python
# for landscapes with less than 1% active variants with boosting, ridge, alde boosting and DNN
plot_mlde_dict(
    get_model_comp_dict(
        mlde_csv,
        lib_stat[~(lib_stat["percent_active"] >= active_cutoff)]["lib"].tolist(),
        n_top=96,
        alde_dir=alde_dir,
    ),
    fig_name="s22",
    n_top=96,
    n_corr=384,
    ifvline=False,
    fig_dir="figs",
    ifsave=True,
)
```
