A notebook for regenerating plot on encoding strategies for MLDE performance

```python
import pandas as pd

from SSMuLA.mlde_analysis import plot_mlde_emb
%matplotlib inline
```

```python
mlde_df = pd.read_csv("results/mlde/all_results.csv")
lib_stat = pd.read_csv("results/landscape/lib_stats.csv")
active_cutoff = 1
```

```python
mlde = mlde_df[
            (mlde_df["n_mut_cutoff"] == "all")
            & (mlde_df["zs"] == "none")
            & (mlde_df["model"] == "boosting")
        ]
mlde
```

```python
mlde_ev = mlde_df[
            (mlde_df["n_mut_cutoff"] == "all")
            & (mlde_df["zs"] == "ev_score")
            & (mlde_df["model"] == "boosting")
        ]
mlde_ev
```

```python
# active no focused trainign
plot_mlde_emb(
    sliced_mlde=mlde,
    lib_list=lib_stat[(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="none",
    fig_name="s21",
    # n_top=96,
    # fig_dir="figs",
    # ifsave=True,
)
```

```python
# active with ev score
plot_mlde_emb(
    sliced_mlde=mlde_ev,
    lib_list=lib_stat[(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="ev_score",
    fig_name="s22",
    # n_top=96,
    # mlde_color_list=["gray", "yellow", "blue", "green"],
    # fig_dir="fig",
    # ifsave=True,
)
```

```python
# <1% active no focused training
plot_mlde_emb(
    sliced_mlde=mlde,
    lib_list=lib_stat[~(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="none",
    fig_name="s23",
    # n_top=96,
    # fig_dir="fig",
    # ifsave=True,
)
```

```python
# <1% active with ev score
plot_mlde_emb(
    sliced_mlde=mlde_ev,
    lib_list=lib_stat[~(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="ev_score",
    fig_name="s24",
    # n_top=96,
    # fig_dir="fig",
    # ifsave=True,
)
```
