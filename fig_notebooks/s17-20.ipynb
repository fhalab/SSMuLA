A notebook for regenerating plot for figure S17 to S20 on encoding strategies for MLDE performance

```python
import pandas as pd 
import numpy as np

from SSMuLA.mlde_analysis import plot_mlde_emb
%matplotlib inline
```

```python
emb_df = pd.read_csv("/disk2/fli/SSMuLA/results4upload/mlde/emb_test.csv")
```

```python
lib_stat = pd.read_csv("/disk2/fli/SSMuLA/results4upload/landscape/lib_stats.csv")
active_cutoff = 1
```

```python
mlde = emb_df[
            (emb_df["n_mut_cutoff"] == "all")
            & (emb_df["zs"] == "none")
            & (emb_df["n_top"] == 96)
            & (emb_df["rep"].isin(np.arange(50)))
            & (emb_df["model"] == "boosting")
        ]
mlde
```

```python
mlde_ev = emb_df[
            (emb_df["n_mut_cutoff"] == "all")
            & (emb_df["zs"] == "ev_score")
            & (emb_df["n_top"] == 96)
            & (emb_df["rep"].isin(np.arange(50)))
            & (emb_df["model"] == "boosting")
        ]
mlde_ev
```

```python
# active no focused trainign
plot_mlde_emb(
    sliced_mlde=mlde,
    lib_list=lib_stat[(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="none",
    fig_name="s17",
    # n_top=96,
    # fig_dir="figs",
    # ifsave=True,
)
```

```python
# active with ev score
plot_mlde_emb(
    sliced_mlde=mlde_ev,
    lib_list=lib_stat[(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="ev_score",
    fig_name="s18",
    # n_top=96,
    # mlde_color_list=["gray", "yellow", "blue", "green"],
    # fig_dir="fig",
    # ifsave=True,
)
```

```python
# <1% active no focused training
plot_mlde_emb(
    sliced_mlde=mlde,
    lib_list=lib_stat[~(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="none",
    fig_name="s19",
    # n_top=96,
    # fig_dir="fig",
    # ifsave=True,
)
```

```python
# <1% active with ev score
plot_mlde_emb(
    sliced_mlde=mlde_ev,
    lib_list=lib_stat[~(lib_stat["percent_active"]>=active_cutoff)]["lib"].tolist(),
    zs="ev_score",
    fig_name="s20",
    # n_top=96,
    # fig_dir="fig",
    # ifsave=True,
)
```
